## Set
임의의 원소를 중복 없이, 순서 없이 담는 집합형 자료구조

## Python의 기본 자료구조 Set을 모방한 클래스 Set을 구현하세요
- __생성자(\_\_init\_\_)__  
list를 받아 중복 제거; 매개 변수 없이 생성 시, 빈 집합 상태로 생성  
- __add(elem)__  
Set에 elem이 존재하지 않으면 추가  
- __discard(elem)__  
Set에 elem이 존재하면 삭제  
- __clear()__  
Set에 존재하는 모든 원소 삭제  
- __\_\_len\_\_()__  
Set에 존재하는 원소 개수 반환  
- __\_\_str\_\_()__  
Set에 존재하는 원소를 '{1, 2, 3}'의 형태로 반환  
- __\_\_constains\_\_(elem)__  
Set에 elem이 존재하면 참 반환, 아니면 거짓 반환  
- __self <= other__  
self가 other의 부분집합이면 참 반환, 아니면 거짓 반환  
- __self >= other__  
other가 self의 부분집합이면 참 반환, 아니면 거짓 반환  
- __self | other__  
self와 other의 원소를 모두 포함하는 합집합 Set 반환  
- __self & other__  
self와 other가 공통으로 포함하는 원소를 포함하는 교집합 Set 반환  
- __self - other__  
self의 원소 중 other에 없는 원소만을 포함하는 차집합 Set 반환  
- __|=, &=, -=__  
위의 기능에 맞추어 구현, In-place operation  
자신을 return 하는 것을 잊지 말 것

```python
class Set:
    def __init__(self, elem = []):
        self.m_list = []
        for i in elem:
            self.add(i)
        
    def add(self, elem):
        if elem not in self.m_list:
            self.m_list.append(elem)
            
    def discard(self, elem):
        if elem in self.m_list:
            self.m_list.remove(elem)
        
    def clear(self):
        self.m_list.clear()
#         self.m_list = []
        
    def __len__(self):
        return len(self.m_list)
        
    def __str__(self):
        return "{" + ", ".join(map(str,self.m_list)) + "}"
    
    def __contains__(self, elem):
        return elem in self.m_list
    
    def __le__(self, other): #self가 부분집합?
        for elem in self.m_list:
            if elem not in other:
                return False
        return True
    
    def __ge__(self, other): #other이 부분집합?
        for elem in other.m_list:
            if elem not in self:
                return False
        return True
    
    def __or__(self, other): #합집합
        new = self.m_list[:]
        for elem in other.m_list:
            new.append(elem)
        return Set(new)
    
    def __and__(self, other): #공통요소
        new = []
        for elem in self.m_list:
            if elem in other.m_list:
                new.append(elem)
        return Set(new)
    
    def __sub__(self, other): #other에 없는 원소
        new = []
        for elem in self.m_list:
            if elem not in other:
                new.append(elem)
        return Set(new)
    
    def __ior__(self, other):
        for elem in other.m_list:
            if elem not in self.m_list:
                self.add(elem)
        return self
    
    def __iand__(self, other):
        temp = self.m_list[:]
        for elem in temp:
            if elem not in other:
                self.discard(elem)
        return self
    
    def __isub__(self, other):
        for elem in other.m_list:
            if elem in self.m_list:
                self.discard(elem)
        return self
```

```python
a = Set([1, 2, 3, 4])
b = Set([1, 2, 3, 4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b. clear()
print(b)
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
